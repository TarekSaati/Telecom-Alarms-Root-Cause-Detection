## Neteco Alarm Names Parsing

```python
neteco_df = pd.read_csv('./Data/Netceo_data_28.csv');
neteco_df.sort_values(by='Alarm Serial Number', ascending=False, inplace=True)
neteco_df['Name'].drop_duplicates().to_csv('neteco_all_names.csv')
```

```python
proc_df = pd.read_csv('neteco_all_names.csv')
abbrev_df = pd.read_csv('neteco_abbrev.csv')
def parse(row):
    for word in row.split():
        if np.sum(abbrev_df['Abbrev'].str.contains(word)):
            id = np.argmax(abbrev_df['Abbrev'].str.contains(word))
            row=row.replace(word, abbrev_df['Meaning'].iloc[id])
    return row
proc_df['Parsed Name'] = proc_df['Name'].apply(parse)
#proc_df['Title'] = proc_df['Parsed Name']
proc_df.to_csv('neteco_all_names.csv')
```

## NCE Alarm Names Parsing

```python
nce_df = pd.read_csv('./Data/NCE_data_28.csv');
nce_df.sort_values(by='Alarm Serial Number', ascending=False, inplace=True)
nce_df['Name'].drop_duplicates().to_csv('nce_all_names.csv')
```

```python
proc_df = pd.read_csv('./Data/nce_all_names.csv')
abbrev_df = pd.read_csv('./Data/nce_abbrev.csv')
proc_df['Parsed Name'] = proc_df['Name'].copy()
def parse(row):
    row = row.lower()
    searchList = [row]
    name = ' '.join(row.split('_')).strip()
    name = ' '.join(name.split('-')).strip()
    name = ' '.join(name.split('(')).strip()
    name = ' '.join(name.split(')')).strip()
    searchList.extend(name.split())
    for word in searchList:
        abbr = np.argmax(abbrev_df['Abbrev'].str.lower().str.contains('^'+word+'$', regex=True))
        if abbr and abbrev_df['done'].iloc[abbr]:
            name=name.replace(word, str(abbrev_df['Meaning'].iloc[abbr]))
    return name
proc_df['Parsed Name'] = proc_df['Name'].apply(parse)
proc_df['Title'] = proc_df['Parsed Name']
proc_df.to_csv('nce_all_names.csv')
```

## MAE Alarms Name Parsing

#### Total Abbrev Loss across 3 systems was 24/460 or 5%

```python
mae_df = pd.read_csv('./Data/MAE_data_28.csv');
mae_df.sort_values(by='Log Serial Number', ascending=False, inplace=True)
mae_df['Name'].drop_duplicates().to_csv('mae_all_names.csv')
```

```python
print('There are ', mae_df['Name'].nunique(), 'Unique titles in MAE system')
```

```python
proc_df = pd.read_csv('./Data/mae_all_names.csv')
abbrev_df = pd.read_csv('./Data/mae_abbrev.csv')
proc_df['Parsed Title'] = proc_df['Title'].copy()
print(len(abbrev_df))
def parse(row):
    global abbrev_df
    #row = row.lower()
    searchList = [row]
    name = ' '.join(row.split('_')).strip()
    name = ' '.join(name.split('-')).strip()
    name = ' '.join(name.split('(')).strip()
    name = ' '.join(name.split(')')).strip()
    searchList.extend(name.split())
    
    for word in searchList:
        if re.findall('[a-z]$', word) or re.findall('^\d$', word) or len(word) < 2 or ' ' in word:
            continue
        abbr = np.argmax(abbrev_df['Abbrev'].str.contains('^'+word+'$', regex=True))
        if not abbr:
            new_abbrev = pd.DataFrame([[word, '', 0]], columns=abbrev_df.columns)
            abbrev_df = pd.concat([abbrev_df, new_abbrev])
        if abbr and abbrev_df['done'].iloc[abbr]:
            name=name.replace(word, str(abbrev_df['Meaning'].iloc[abbr]))
    return name
proc_df['Parsed Title'] = proc_df['Title'].apply(parse)
print(len(abbrev_df))
abbrev_df.to_csv('mae_abbrev.csv')
proc_df.to_csv('mae_all_names.csv')
```

## Triplet Dataset creation

```python
nce_df = pd.read_csv('./Data/neteco_all_names.csv')
n_titles = len(nce_df)
label = ['Category', 'Level']
nce_triplets = []

for i in range(n_titles):    
    
    query = nce_df.iloc[i]['Title']
    query_label = nce_df[label].iloc[i]
    cond = nce_df[label] == query_label
    rel_cond = np.logical_and.reduce(cond, axis=1)
    unrel_cond = ~np.logical_or.reduce(cond, axis=1)
    related = nce_df[rel_cond].sample()['Title'].iloc[0]
    unrelated = nce_df[unrel_cond].sample()['Title'].iloc[0]
    nce_triplets.append([query, related, unrelated])
```
